```python
amostra = dados_ibovespa[["date", "close"]].tail(1000).sort_values(by="date",ascending=False).reset_index()
amostra.head()
```

```python
time_window = 7
```

```python
train_y = list()
train_X = list()

for idx in range(amostra.shape[0]-time_window):
    date = amostra['date'][idx]
    y = [amostra['close'][idx]]
    x = list(amostra['close'][(idx+1):(time_window+idx+1)].values)
    train_y.append(y)
    train_X.append(x)   
```

```python
# Turn the input pandas dataframe into tensors
train_sample_y = Variable(torch.Tensor(np.array(train_y))).float().squeeze()
train_sample_X = Variable(torch.Tensor(np.array(train_X))).float()

# Build the dataset
train_sample_ds = torch.utils.data.TensorDataset(train_sample_X, train_sample_y)
# Build the dataloader
train_loader = torch.utils.data.DataLoader(train_sample_ds, batch_size=50)
```

```python
class LSTM(nn.Module):
    def __init__(self, input_size=time_window, hidden_layer_size=10, output_size=1):
        super().__init__()
        self.hidden_layer_size = hidden_layer_size

        self.lstm = nn.LSTM(input_size, hidden_layer_size)

        self.linear = nn.Linear(hidden_layer_size, output_size)

        self.hidden_cell = (torch.zeros(1,1,self.hidden_layer_size),
                            torch.zeros(1,1,self.hidden_layer_size))
        
    def init_hidden(self):
        return (
                Variable(torch.zeros(1,1,self.hidden_layer_size)),
                Variable(torch.zeros(1,1,self.hidden_layer_size))
                )

    def forward(self, input_seq):
        lstm_out, self.hidden_cell = self.lstm(input_seq.view(len(input_seq) ,1, -1), self.hidden_cell)
        predictions = self.linear(lstm_out.view(len(input_seq), -1))
        return predictions[-1]
```

```python
def train(model, train_loader, epochs, optimizer, loss_fn, device):
    for epoch in tqdm(range(1, epochs + 1)):
        model.train()
        total_loss = 0
        for batch in train_loader:         
            batch_X, batch_y = batch
            
            batch_X = batch_X.to(device)
            batch_y = batch_y.to(device)
            # TODO: Complete this train method to train the model provided.
            
            output = model(batch_X)
            loss = loss_fn(output, batch_y)
            model.hidden = model.init_hidden()
            loss.backward(retain_graph=True)
            optimizer.step()
            optimizer.zero_grad()
            
            ###
            
            total_loss += loss.data.item()
        print("Epoch: {}, BCELoss: {}".format(epoch, total_loss / len(train_loader)))
```

```python
epochs = 100
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = LSTM().to(device)
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

train(model, train_loader, epochs, optimizer, loss_fn, device)
```
